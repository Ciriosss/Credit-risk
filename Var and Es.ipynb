```python
import scipy.stats as stats


mean = 1
std_dev = 3
confidence_level = 0.95

# Calculate the z-score
z_score = stats.norm.ppf(confidence_level)

# Calculate VaR
var_95 = mean + z_score * std_dev

es_95 = mean + std_dev * stats.norm.pdf(z_score) / (1 - confidence_level)

# Display the result
print(f"VaR at {confidence_level * 100}% confidence level for 1 year: {var_95:.4f} million euros")



print(var_99)
print(stats.norm.pdf(z_score))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


mean = 1
std_dev = 3
confidence_level = 0.99


z_score = stats.norm.ppf(confidence_level)


var_99 = mean + z_score * std_dev


es_99 = mean + std_dev * stats.norm.pdf(z_score) / (1 - confidence_level)


x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)


pdf_values = stats.norm.pdf(x_values, loc=mean, scale=std_dev)


plt.plot(x_values, pdf_values, label='Normal Distribution')


plt.axvline(x=var_99, color='r', linestyle='--', label=f'VaR at {confidence_level * 100}%')


plt.axvline(x=es_99, color='g', linestyle='--', label=f'ES at {confidence_level * 100}%')

# Add labels and legend
plt.xlabel('Portfolio Loss (million euros)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution with VaR and ES')
plt.legend()

# Show the plot
plt.show()

```
